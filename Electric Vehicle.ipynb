```python
df = pd.DataFrame(data)
```

```python
Quantidade = df.shape[0]
```

```python
print("Quantidade de linhas:", Quantidade)
```

```python
QuantidadeFabricante = df['Fabricante'].value_counts()
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns
import plotly.express as px
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
%matplotlib inline
```

```python
dez_maiores_fabricantes = QuantidadeFabricante.nlargest(10)
plt.figure(figsize=(10, 6), dpi=150)
plt.bar(dez_maiores_fabricantes.index, dez_maiores_fabricantes.values)
plt.title('Quantidade de Carros Vendidos dos 10 Maiores Fabricantes')
plt.ylabel('QuantidadeFabricante')
plt.xlabel('Fabricante')
plt.xticks(rotation=45)
plt.show()
```

```python
X = data.drop("Fabricante", axis=1)
y = data["Fabricante"]
```

```python
X_all_columns = data.drop("Fabricante", axis=1).copy()
X_all_columns["Fabricante"] = data["Fabricante"]
```

```python
X_encoded = pd.get_dummies(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.3, random_state=42)
```

```python
n_estimators = 10
random_forest = RandomForestClassifier(n_estimators=n_estimators, random_state=42)
random_forest.fit(X_train, y_train)
```

```python
y_pred = random_forest.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Acurcia:", accuracy)

print("Relatrio de Classificao:")
print(classification_report(y_test, y_pred))
```

```python
probabilities = random_forest.predict_proba(X_test)

# Encontre o ndice correspondente  coluna "Tesla" no conjunto de classes do modelo
fabricante_index = random_forest.classes_.tolist().index("TESLA")

# Exiba a probabilidade de cada carro eltrico no conjunto de teste ser da marca "Tesla"
average_probability_fabricante = probabilities[:, fabricante_index].mean()

print("Probabilidade mdia de um carro ser da fabricante 'TESLA':", average_probability_fabricante)
```
